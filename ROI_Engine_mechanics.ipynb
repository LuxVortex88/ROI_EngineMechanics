```python
import numpy as np
import matplotlib.pyplot as plt

# Time in months
months = np.arange(0, 60)

# Upgrade types and costs
upgrades = {
    'Turbocharged': {'cost': 8000, 'monthly_gain': 300},
    'Hybrid System': {'cost': 12000, 'monthly_gain': 450},
    'Thermal Coating': {'cost': 5000, 'monthly_gain': 180}
}

plt.figure(figsize=(12, 6))

for name, data in upgrades.items():
    gain = data['monthly_gain'] * months
    roi = (gain - data['cost']) / data['cost'] * 100
    plt.plot(months, roi, label=f'{name} ROI')

# Plot formatting
plt.axhline(0, color='gray', linestyle='--')
plt.title('ROI on Engine Mechanical Upgrades')
plt.xlabel('Time (months)')
plt.ylabel('ROI (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
